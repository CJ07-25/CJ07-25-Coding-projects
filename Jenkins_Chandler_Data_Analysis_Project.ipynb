```python
import pandas as pd
df = pd.read_csv('company_sales_data.csv')
print(df)

```

```python

from pylab import plt, mpl
plt.style.use('seaborn-v0_8')

print(df.columns)


plt.plot(df['month_number'], df['total_profit'],marker='o',linestyle='--',linewidth='3',label='Profit Data of Last Year', color=('r'))
plt.xlabel('Month')
plt.ylabel('Profit in Dollars')
plt.title('Company Sales Data of Last Year')
plt.legend(loc='upper center', bbox_to_anchor=(0.8, 0.11), ncol=1)
plt.show()


```

```python
plt.scatter(df['month_number'], df['toothpaste'],marker='o',label='Toothpaste Sales Data', color=('r'))
print(df.columns)
plt.xlabel('Month')
plt.ylabel('number of units sold')
plt.title('Toothpaste Sells Data')
plt.legend(loc='upper center', bbox_to_anchor=(0.8, 0.11), ncol=1)
plt.grid(linestyle = '--')
plt.show()
```

```python
import matplotlib.pyplot as plt



# Display column names
print(df.columns)

# Sum total units for each item
items = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
units_per_item = df[items].sum()
# Correct the pie chart code'
# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(units_per_item, labels=items, autopct='%1.1f%%', startangle=360)
plt.title('Sales Data')
plt.legend(items, title='Products', loc='lower right')
plt.show()
plt.show()
```
