```python
import seaborn as sns
sns.set_style('whitegrid')
# built-in tips dataset
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.jointplot(x='fare', y='age', data=titanic, kind='scatter')
```

```python
sns.histplot(data=titanic, x='fare', color='red', kde=False)
```

```python
sns.boxplot(data=titanic, x='class', y='age')
```

```python
sns.swarmplot(data=titanic, x='class', y='age', hue='class',)
```

```python
sns.countplot(x='sex', data=titanic)
```

```python
# Select relevant columns
aj = titanic[['survived', 'pclass', 'age', 'sibsp', 'parch', 'fare', 'adult_male', 'alone']]
```
